```python
import numpy as np

eps = 1e-9

cols = ["AMT_CREDIT", "AMT_GOODS_PRICE", "AMT_ANNUITY", "AMT_APPLICATION", "CNT_PAYMENT"]
for c in cols:
    if c in df.columns:
        df[c] = df[c].fillna(df[c].median())   # <-- ØªØµØ­ÙŠØ­ Ù…Ù‡Ù…

df["CREDIT_GOODS_RATIO"] = df["AMT_CREDIT"] / (df["AMT_GOODS_PRICE"] + eps)
df["ANNUITY_CREDIT_RATIO"] = df["AMT_ANNUITY"] / (df["AMT_CREDIT"] + eps)
df["APPLICATION_CREDIT_RATIO"] = df["AMT_APPLICATION"] / (df["AMT_CREDIT"] + eps)
df["ANNUITY_APPLICATION_RATIO"] = df["AMT_ANNUITY"] / (df["AMT_APPLICATION"] + eps)
df["INSTALMENT_CREDIT_RATIO"] = df["CNT_PAYMENT"] / (df["AMT_CREDIT"] + eps)

ratio_cols = [
    "CREDIT_GOODS_RATIO",
    "ANNUITY_CREDIT_RATIO",
    "APPLICATION_CREDIT_RATIO",
    "ANNUITY_APPLICATION_RATIO",
    "INSTALMENT_CREDIT_RATIO"
]

for c in ratio_cols:
    low = df[c].quantile(0.005)
    high = df[c].quantile(0.995)
    df[c] = df[c].clip(low, high)
```

```python
import numpy as np
import pandas as pd
from numba import njit

@njit
def rolling_mean_last_x_days(days, credit, window):
    n = len(days)
    result = np.zeros(n)
    for i in range(n):
        d0 = days[i]
        total = 0.0
        count = 0
        for j in range(i, -1, -1):
            diff = d0 - days[j]
            if diff <= window and diff >= 0:
                total += credit[j]
                count += 1
            else:
                break
        result[i] = total / count
    return result


df = df.sort_values(["SK_ID_CURR", "DAYS_DECISION"])

df["MEAN_CREDIT_3M"] = np.nan  # Ø¨Ù†Ø­Ø³Ø¨ Ù…ØªÙˆØ³Ø· Ù…Ø¨Ù„Øº Ø§Ù„Ù‚Ø±ÙˆØ¶ Ø®Ù„Ø§Ù„ Ø¢Ø®Ø± 90 ÙŠÙˆÙ… Ù„Ù†ÙØ³ Ø§Ù„Ø¹Ù…ÙŠÙ„ 

for cust_id, group in df.groupby("SK_ID_CURR"):
    idx = group.index
    days = group["DAYS_DECISION"].values.astype(np.int64)
    credit = group["AMT_CREDIT"].values.astype(np.float64)

    df.loc[idx, "MEAN_CREDIT_3M"] = rolling_mean_last_x_days(days, credit, 90)
eps = 1e-9

df["CREDIT_3M_TO_LAST_RATIO"] = df["MEAN_CREDIT_3M"] / (df["AMT_CREDIT"] + eps)

# Ø¨Ù†Ù‚Ø³Ù‘Ù… Ø§Ù„Ù…ØªÙˆØ³Ø· Ø¯Ù‡ Ø¹Ù„Ù‰ Ù‚ÙŠÙ…Ø© Ø§Ù„Ù‚Ø±Ø¶ Ø§Ù„Ø­Ø§Ù„ÙŠ 
# Ø¹Ù„Ø´Ø§Ù† Ù†Ø¹Ø±Ù Ø§Ù„Ù‚Ø±Ø¶ Ø§Ù„Ø­Ø§Ù„ÙŠ ÙƒØ¨ÙŠØ± ÙˆÙ„Ø§ ØµØºÙŠØ± Ù…Ù‚Ø§Ø±Ù†Ø© Ø¨Ø§Ù„Ù‚Ø±ÙˆØ¶ Ø§Ù„Ù„ÙŠ Ø£Ø®Ø¯Ù‡Ø§ ÙÙŠ Ø¢Ø®Ø± 3 Ø´Ù‡ÙˆØ± 

df.drop(columns=["MEAN_CREDIT_3M"], inplace=True, errors="ignore")
```
