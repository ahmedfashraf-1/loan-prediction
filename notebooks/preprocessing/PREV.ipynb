```python
def fix_portfolio(row):
    current = str(row['NAME_PORTFOLIO']).strip()
    if current != 'XNA':
        return current 

    product = str(row['PRODUCT_COMBINATION']).lower().strip() if pd.notna(row['PRODUCT_COMBINATION']) else ''
    industry = str(row['NAME_SELLER_INDUSTRY']).lower().strip() if pd.notna(row['NAME_SELLER_INDUSTRY']) else ''

    if product.startswith('cash'):
        return 'Cash'
    elif product.startswith('card'):
        return 'Cards'
    elif product.startswith('pos'):
        if 'industry' in industry:
            return 'Cars'
        else:
            return 'POS'
    else:
        return current

df['NAME_PORTFOLIO'] = df.apply(fix_portfolio, axis=1)
```

##### <span style="color:#a8dbc9d1;">Fill XNA in `NAME_YILED_GROUP` using NAME_CONTRACT_TYPE,PRODUCT_COMBINATION,NAME_PORTFOLIO

```python
def classify_yield_group_final(row):
    # --- Extract & clean features ---
    contract = str(row.get('NAME_CONTRACT_TYPE', '')).lower()
    product = str(row.get('PRODUCT_COMBINATION', '')).lower()
    portfolio = str(row.get('NAME_PORTFOLIO', '')).lower()
    client_type = str(row.get('NAME_CLIENT_TYPE', '')).lower()
    channel = str(row.get('CHANNEL_TYPE', '')).lower()

    annuity = row.get('AMT_ANNUITY')
    annuity = float(annuity) if annuity not in [None, ''] else 0.0

    payments = row.get('CNT_PAYMENT')
    payments = int(payments) if payments not in [None, ''] else 999

    if 'high' in product:
        return 'High'
    elif 'middle' in product:
        return 'Middle'
    elif 'low' in product:
        return 'Low'

    if 'with' in product:
        if payments <= 12:
            return 'High'
        elif payments > 36:
            return 'Low'
        else:
            return 'Middle'
    elif 'without' in product:
        if annuity > 25000:
            return 'Middle'
        else:
            return 'Low'

    if 'revolving' in contract:
        if client_type == 'new':
            return 'High'
        elif client_type in ('repeater', 'refreshed'):
            if portfolio == 'pos':
                return 'Middle'
            elif portfolio == 'cash':
                return 'Low'
            else:
                return 'Middle'
        else:
            if payments <= 12:
                return 'High'
            elif payments <= 24:
                return 'Middle'
            else:
                return 'Low'

    if 'pos' in portfolio:
        return 'Middle'
    elif 'cash' in portfolio:
        if annuity > 30000:
            return 'Middle'
        else:
            return 'Low'
    elif 'cards' in portfolio:
        return 'High'

    if 'credit and cash' in channel:
        return 'High'
    elif 'country-wide' in channel:
        return 'Middle'
    elif 'contact center' in channel:
        return 'Low'

    return 'Unknown'
```

```python
df['NAME_YIELD_GROUP'] = df.apply(classify_yield_group_final, axis=1)
```
