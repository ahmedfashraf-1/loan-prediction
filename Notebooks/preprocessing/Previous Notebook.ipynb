```python
def fix_portfolio(row):
    current = str(row['NAME_PORTFOLIO']).strip()
    if current != 'XNA':
        return current 

    product = str(row['PRODUCT_COMBINATION']).lower().strip() if pd.notna(row['PRODUCT_COMBINATION']) else ''
    industry = str(row['NAME_SELLER_INDUSTRY']).lower().strip() if pd.notna(row['NAME_SELLER_INDUSTRY']) else ''

    if product.startswith('cash'):
        return 'Cash'
    elif product.startswith('card'):
        return 'Cards'
    elif product.startswith('pos'):
        if 'industry' in industry:
            return 'Cars'
        else:
            return 'POS'
    else:
        return current

df['NAME_PORTFOLIO'] = df.apply(fix_portfolio, axis=1)
```

```python
med_low  = df[df['NAME_YIELD_GROUP'] == 'low']['AMT_ANNUITY'].median()
med_mid  = df[df['NAME_YIELD_GROUP'] == 'middle']['AMT_ANNUITY'].median()
med_high = df[df['NAME_YIELD_GROUP'] == 'high']['AMT_ANNUITY'].median()

print("Median Low:   ", med_low)
print("Median Middle:", med_mid)
print("Median High:  ", med_high)


# ==========================================
# 2) ØªØ­Ø¯ÙŠØ¯ thresholds Ø¨Ù†Ø§Ø¡Ù‹ Ø¹Ù„Ù‰ Ø§Ù„Ù…ÙŠØ¯ÙŠØ§Ù†
# ==========================================

t1 = (med_low + med_mid) / 2
t2 = (med_mid + med_high) / 2

print("\nThreshold 1 (Lowâ†’Middle): ", t1)
print("Threshold 2 (Middleâ†’High):", t2)


# ==============================================
# 3) Ø¯Ø§Ù„Ø© Ø¨Ø³ÙŠØ·Ø© Ù„ØªØµÙ†ÙŠÙ XNA ÙÙ‚Ø· Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… annuity
# ==============================================

def fill_yield(ann):
    if pd.isna(ann):
        return 'Unknown'   # Ù„Ùˆ Ù…ÙÙŠØ´ Ø£Ù†ÙŠÙˆØªÙŠ Ø£ØµÙ„Ø§Ù‹
    if ann < t1:
        return 'low'
    elif ann < t2:
        return 'middle'
    else:
        return 'high'


# ===================================
# 4) ØªØ·Ø¨ÙŠÙ‚ Ø§Ù„Ø¯Ø§Ù„Ø© Ø¹Ù„Ù‰ XNA ÙÙ‚Ø·
# ===================================

mask = df['NAME_YIELD_GROUP'] == 'XNA'
df.loc[mask, 'NAME_YIELD_GROUP'] = df.loc[mask, 'AMT_ANNUITY'].apply(fill_yield)
```

```python
import numpy as np

eps = 1e-9

cols = ["AMT_CREDIT", "AMT_GOODS_PRICE", "AMT_ANNUITY", "AMT_APPLICATION", "CNT_PAYMENT"]
for c in cols:
    if c in df.columns:
        df[c] = df[c].fillna(df[c].median())   # <-- ØªØµØ­ÙŠØ­ Ù…Ù‡Ù…

df["CREDIT_GOODS_RATIO"] = df["AMT_CREDIT"] / (df["AMT_GOODS_PRICE"] + eps)
df["ANNUITY_CREDIT_RATIO"] = df["AMT_ANNUITY"] / (df["AMT_CREDIT"] + eps)
df["APPLICATION_CREDIT_RATIO"] = df["AMT_APPLICATION"] / (df["AMT_CREDIT"] + eps)
df["ANNUITY_APPLICATION_RATIO"] = df["AMT_ANNUITY"] / (df["AMT_APPLICATION"] + eps)
df["INSTALMENT_CREDIT_RATIO"] = df["CNT_PAYMENT"] / (df["AMT_CREDIT"] + eps)

ratio_cols = [
    "CREDIT_GOODS_RATIO",
    "ANNUITY_CREDIT_RATIO",
    "APPLICATION_CREDIT_RATIO",
    "ANNUITY_APPLICATION_RATIO",
    "INSTALMENT_CREDIT_RATIO"
]

for c in ratio_cols:
    low = df[c].quantile(0.005)
    high = df[c].quantile(0.995)
    df[c] = df[c].clip(low, high)
```

```python
import numpy as np
import pandas as pd
from numba import njit

@njit
def rolling_mean_last_x_days(days, credit, window):
    n = len(days)
    result = np.zeros(n)
    for i in range(n):
        d0 = days[i]
        total = 0.0
        count = 0
        for j in range(i, -1, -1):
            diff = d0 - days[j]
            if diff <= window and diff >= 0:
                total += credit[j]
                count += 1
            else:
                break
        result[i] = total / count
    return result


df = df.sort_values(["SK_ID_CURR", "DAYS_DECISION"])

df["MEAN_CREDIT_3M"] = np.nan  # Ø¨Ù†Ø­Ø³Ø¨ Ù…ØªÙˆØ³Ø· Ù…Ø¨Ù„Øº Ø§Ù„Ù‚Ø±ÙˆØ¶ Ø®Ù„Ø§Ù„ Ø¢Ø®Ø± 90 ÙŠÙˆÙ… Ù„Ù†ÙØ³ Ø§Ù„Ø¹Ù…ÙŠÙ„ 

for cust_id, group in df.groupby("SK_ID_CURR"):
    idx = group.index
    days = group["DAYS_DECISION"].values.astype(np.int64)
    credit = group["AMT_CREDIT"].values.astype(np.float64)

    df.loc[idx, "MEAN_CREDIT_3M"] = rolling_mean_last_x_days(days, credit, 90)
eps = 1e-9

df["CREDIT_3M_TO_LAST_RATIO"] = df["MEAN_CREDIT_3M"] / (df["AMT_CREDIT"] + eps)

# Ø¨Ù†Ù‚Ø³Ù‘Ù… Ø§Ù„Ù…ØªÙˆØ³Ø· Ø¯Ù‡ Ø¹Ù„Ù‰ Ù‚ÙŠÙ…Ø© Ø§Ù„Ù‚Ø±Ø¶ Ø§Ù„Ø­Ø§Ù„ÙŠ 
# Ø¹Ù„Ø´Ø§Ù† Ù†Ø¹Ø±Ù Ø§Ù„Ù‚Ø±Ø¶ Ø§Ù„Ø­Ø§Ù„ÙŠ ÙƒØ¨ÙŠØ± ÙˆÙ„Ø§ ØµØºÙŠØ± Ù…Ù‚Ø§Ø±Ù†Ø© Ø¨Ø§Ù„Ù‚Ø±ÙˆØ¶ Ø§Ù„Ù„ÙŠ Ø£Ø®Ø¯Ù‡Ø§ ÙÙŠ Ø¢Ø®Ø± 3 Ø´Ù‡ÙˆØ± 

df.drop(columns=["MEAN_CREDIT_3M"], inplace=True, errors="ignore")
```
