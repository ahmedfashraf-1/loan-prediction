```python
import  numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("final_merged_data.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include="object").columns.tolist()
# Label Encoding
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))

print("Done! Encoded columns:")
print(cat_cols)
```

```python
df.head()
```

```python
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(15, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix (Numeric Columns Only)")
plt.show()
```

```python
import numpy as np

# Ø§Ø­Ø³Ø¨ Correlation Matrix
corr_matrix = df.corr().abs()

# Ø®Ù„ÙŠ diagonal = 0 Ø¹Ø´Ø§Ù† Ù…ØªØ§Ø®Ø¯Ø´ Ø§Ø±ØªØ¨Ø§Ø· Ø§Ù„Ø¹Ù…ÙˆØ¯ Ù…Ø¹ Ù†ÙØ³Ù‡
np.fill_diagonal(corr_matrix.values, 0)

# Ø­Ø¯ Ø§Ù„Ø§Ø±ØªØ¨Ø§Ø· (0.90 Ù…Ø«Ù„Ø§Ù‹)
threshold = 0.90

# Ø´ÙˆÙ ÙƒÙ„ Ø§Ù„Ø£Ø²ÙˆØ§Ø¬ Ø§Ù„Ù„ÙŠ Ø¨ÙŠÙ†Ù‡Ù… Ø§Ø±ØªØ¨Ø§Ø· Ø¹Ø§Ù„ÙŠ
high_corr_pairs = [
    (col1, col2, corr_matrix.loc[col1, col2])
    for col1 in corr_matrix.columns
    for col2 in corr_matrix.columns
    if col1 < col2 and corr_matrix.loc[col1, col2] > threshold
]

# Ø§Ø·Ø¨Ø¹Ù‡Ù…
for c1, c2, corr in high_corr_pairs:
    print(f"{c1}  <-->  {c2}   | corr = {corr:.3f}")
```
